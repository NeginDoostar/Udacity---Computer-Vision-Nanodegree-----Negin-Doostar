```python
# Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Path of images
day_image_path = 'day_image.jpg'
night_image_path = 'night_image.jpg'

# Read images
day_image = cv2.imread(day_image_path)
night_image = cv2.imread(night_image_path)

# Convert image to RGB format
rgb_day_image = cv2.cvtColor(day_image, cv2.COLOR_BGR2RGB)
rgb_night_image = cv2.cvtColor(night_image, cv2.COLOR_BGR2RGB)

# Plot images
f, ax = plt.subplots(1, 2)
ax[0].set_title('day')
ax[0].imshow(rgb_day_image)
ax[1].set_title('night')
ax[1].imshow(rgb_night_image)
```

```python
# Convert images to HSV format
hsv_day_image = cv2.cvtColor(day_image, cv2.COLOR_BGR2HSV)
hsv_night_image = cv2.cvtColor(night_image, cv2.COLOR_BGR2HSV)

# Plot day image
hsv_day_image_copy = hsv_day_image.copy()
h_day = hsv_day_image_copy[:, :, 0]
s_day = hsv_day_image_copy[:, :, 1]
v_day = hsv_day_image_copy[:, :, 2]

f, axs = plt.subplots(1, 3)
axs[0].set_title('hue channel')
axs[0].imshow(h_day, cmap = 'gray')
axs[1].set_title('saturation channel')
axs[1].imshow(s_day, cmap = 'gray')
axs[2].set_title('value channel')
axs[2].imshow(v_day, cmap = 'gray')

# Plot night image
# Plot day image
hsv_night_image_copy = hsv_night_image.copy()
h_night = hsv_night_image_copy[:, :, 0]
s_night = hsv_night_image_copy[:, :, 1]
v_night = hsv_night_image_copy[:, :, 2]

f, axs = plt.subplots(1, 3)
axs[0].set_title('hue channel')
axs[0].imshow(h_night, cmap = 'gray')
axs[1].set_title('saturation channel')
axs[1].imshow(s_night, cmap = 'gray')
axs[2].set_title('value channel')
axs[2].imshow(v_night, cmap = 'gray')
```

```python
# Find sum of pixel in value channel

v_day = np.array(v_day)
v_night = np.array(v_night)

sum_day = np.sum(v_day)
sum_night = np.sum(v_night)

print(sum_day)
print(sum_night)
```
