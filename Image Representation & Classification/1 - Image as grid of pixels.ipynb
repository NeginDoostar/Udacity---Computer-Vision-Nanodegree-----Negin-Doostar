## Images as grid of pixels

```python
# libraries
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import cv2
```

```python
# read an image and plot it
image = mpimg.imread('colorfull.jpg')
print(f'image dimantion: ', image.shape)
plt.imshow(image)
```

```python
# change to gray scale and plot it
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_image, cmap = 'gray')
```
