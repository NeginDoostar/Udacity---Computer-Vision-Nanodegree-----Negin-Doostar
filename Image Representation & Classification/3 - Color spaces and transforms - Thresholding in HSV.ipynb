```python
# Libraries

import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# Load an image and plot it to see the result

image = cv2.imread('colorfull.jpg')
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_image)
```

## Plot color channels in RGB

```python
rgb_image_copy = rgb_image.copy()
r = rgb_image_copy[:, :, 0]
g = rgb_image_copy[:, :, 1]
b = rgb_image_copy[:, :, 2]

f, axs = plt.subplots(1, 3)
axs[0].set_title('red channel')
axs[0].imshow(r, cmap = 'gray')
axs[1].set_title('green channel')
axs[1].imshow(g, cmap = 'gray')
axs[2].set_title('blue channel')
axs[2].imshow(b, cmap = 'gray')
```

## Plot Color channels in HSV

```python
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
plt.imshow(hsv_image)
```

```python
hsv_image_copy = hsv_image.copy()
h = hsv_image_copy[:, :, 0]
s = hsv_image_copy[:, :, 1]
v = hsv_image_copy[:, :, 2]

f, axs = plt.subplots(1, 3)
axs[0].set_title('hue channel')
axs[0].imshow(h, cmap = 'gray')
axs[1].set_title('saturation channel')
axs[1].imshow(s, cmap = 'gray')
axs[2].set_title('value channel')
axs[2].imshow(v, cmap = 'gray')
```

## Find red mask in HSV image with hue channel

```python
# Define threshold boundaries to make a red and orange mask

# Lower range of red
lower_red1 = np.array([0, 150, 150])
upper_red1 = np.array([10, 255, 255])

# Upper range of red
lower_red2 = np.array([160, 150, 150])
upper_red2 = np.array([179, 255, 255])

mask1 = cv2.inRange(hsv_image, lower_red1, upper_red1)
mask2 = cv2.inRange(hsv_image, lower_red2, upper_red2)
red_mask = cv2.bitwise_or(mask1, mask2)

# Plot red pixels
rgb_image_copy2 = rgb_image.copy()
rgb_image_copy2 [red_mask == 0] = [0, 0, 0]
plt.imshow(rgb_image_copy2)
```
