## Blue Threshold

```python
# Libraries

import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# Load an image and plot it to see the result

image = cv2.imread('flower.png')
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_image)
```

```python
# Define threshold boundaries to make a blue mask

blue_lower_threshold = np.array([0, 0, 220])
blue_upper_threshold = np.array([110, 110, 255])

blue_mask = cv2.inRange(rgb_image.copy(), blue_lower_threshold, blue_upper_threshold)
plt.imshow(blue_mask, cmap = 'gray')
```

```python
# Show the image with blue_mask

mask_image = rgb_image.copy()
mask_image [blue_mask != 0] = [0, 0, 0]
plt.imshow(mask_image)
```

```python
# Load an image to change blue part of the flower image

back_image = cv2.imread('space.png')
rgb_back_image = cv2.cvtColor(back_image, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_back_image)
```

```python
# Campare the shape of images

print(rgb_image.shape)
print(rgb_back_image.shape)
```

```python
# Crop background of blue 

crop_rgb_back_image = rgb_back_image[0:255, 0:197]
plt.imshow(crop_rgb_back_image)
```

```python
# Find blue mask in crop_rgb_back_image and change the flower image with this 

crop_rgb_back_image[blue_mask == 0] = [0, 0, 0]
new_image = crop_rgb_back_image + mask_image
plt.imshow(new_image)
```
