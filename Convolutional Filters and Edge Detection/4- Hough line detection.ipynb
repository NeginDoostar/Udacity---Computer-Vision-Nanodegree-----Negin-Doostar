```python
# Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read an image
image = cv2.imread('book.jpg', 0)

# Plot the image
plt.imshow(image, cmap = 'gray')
```

```python
# Find edges by canny edge detection
low_threshold, high_threshold = 70, 100
edges = cv2.Canny(image, low_threshold, high_threshold)

# Plot the results
plt.imshow(edges, cmap = 'gray')
```

```python
# Define parameters for hough line detection
rho = 1
theta = np.pi / 180
threshold = 120
min_line_length = 60
max_line_gap = 1

# Hough line detection
line_image = cv2.HoughLinesP(edges, rho, theta, threshold, np.array([]), min_line_length, max_line_gap)

# Iterative over the output "lines" and draw lines on the image copy
line_image_copy = image.copy()

for line in line_image:
    for x1, y1, x2, y2 in line:
        cv2.line(line_image_copy, (x1, y1), (x2, y2), (255, 0, 0), 5)

# Plot the result
plt.imshow(line_image_copy)
```
