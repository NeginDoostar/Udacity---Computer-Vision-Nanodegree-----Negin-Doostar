```python
# Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read an image in grayscale format
image = cv2.imread('day_image.jpg',0)

# Plot the image
plt.imshow(image, cmap = 'gray')
```

```python
# Wide canny filter
wide_lower, wide_upper = 20, 240
wide_filterd_image = cv2.Canny(image, wide_lower, wide_upper)

# Tight canny filter
tight_lower, tight_upper = 180, 240
tight_filterd_image = cv2.Canny(image, tight_lower, tight_upper)

# Plot the result
f, ax = plt.subplots(1, 2)
ax[0].imshow(wide_filterd_image, cmap = 'gray')
ax[1].imshow(tight_filterd_image, cmap = 'gray')
```
